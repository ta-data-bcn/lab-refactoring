## Function to gather message to encrypt and cipher key
I've create a simple function to gather input from the user and stores the values to be used in a the main function to encrypt.

```python
def user_input():
    message_to_encrypt = input("Please insert your secret message ")
    s = int(input("Please insert the key to use for cipher "))
    return message_to_encrypt, s

message, step = user_input()        
```

## Encryption function

```python
def encrypt():
    result = ""      
    # encrypting text 
    for i in range(len(message)): 
        char = message[i]   
        # Encrypt uppercase characters according to ASCII codes
        if (char.isupper()): 
            result += chr((ord(char) + step - 65) % 26 + 65)  
        # Encrypt lowercase characters according to ASCII codes
        else: 
            result += chr((ord(char) + step - 97) % 26 + 97)
    return result 
encrypt()
```

## Gathering text to decrypt

```python
def encrypted_user_input():
    message_to_decrypt = input("Please insert your secret message ")
    key = int(input("Please insert the key to use for decipher "))
    return message_to_decrypt, key

encryption, key = encrypted_user_input()
```

## Decryption function

```python
def decrypt():
    txt_result = ""   
    # decrypting text 
    for i in range(len(encryption)): 
        char = encryption[i]   
        # Encrypt uppercase characters 
        if (char.isupper()): 
            txt_result += chr((ord(char) - key - 65) % 26 + 65)  
        # Encrypt lowercase characters 
        else: 
            txt_result += chr((ord(char) - key - 97) % 26 + 97)   
    return txt_result 
decrypt()
```
